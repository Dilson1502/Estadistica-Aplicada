```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # Permite la generación de gráficos
from mpl_toolkits.mplot3d import Axes3D   # Permite agregar eje tridimensionales
import random                             # Permiten obtener de distintos modos números aleatorios
```

## Escenario 1

El concepto de interacción de interacción es posible representarlo

```python
def f(x1,x2):
    return 35.5 + 10.5*x1 + 5.5*x2 + 0.5*x1*x2
x = np.linspace(-1,1,100)
p = np.linspace(-1,1,100)

X,P = np.meshgrid(x,p)
y = f(X,P)
```

```python
fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection='3d')

ax.plot_surface(X, P, y, rstride=1, cstride=1, cmap='viridis',edgecolor='none');
```

También podemos graficar las interacciones en forma de curvas de nivel mediante un gráfico de contorno

```python
cs = plt.contour(X,P,y,15)
plt.clabel(cs, inline=1, fontsize=8)
plt.show()
```

## Escenario 2
